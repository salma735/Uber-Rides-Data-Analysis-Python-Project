# Data Preprocessing

# TO change NaN To NOT

```python

dataset['PURPOSE'].fillna('NOT', inplace = True)
```

```python
dataset.head()
```

# To change formate of Star_Date and End_Date to Date formate

```python
dataset['START_DATE'] = pd.to_datetime(dataset['START_DATE'], errors = 'coerce')
dataset['END_DATE'] = pd.to_datetime(dataset['END_DATE'], errors = 'coerce')
```

 (errors = coerce is used to put nat (not a time) where date is inappropriate like 26pm,25 am etc.)
                                                                                  

```python
dataset
```

# we will create two column name Date & Time by extracting date and time from column START_DATE

```python
from datetime import datetime

dataset['DATE'] = pd.DatetimeIndex(dataset['START_DATE']).date
dataset['TIME'] = pd.DatetimeIndex(dataset['START_DATE']).hour
```

```python
dataset.head()
```

# we will create a column name DAY_NIGHT by dividing TIME into category(morning,noon,eve, night)

```python
dataset['DAY-NIGHT'] = pd.cut(x = dataset['TIME'], bins = [0,10,15,19,24], labels =['Moorning','Afternoon','Evening','Night'])
```

 The bins parameter is used to define the intervals or "buckets" into which the data in the TIME column will be categorized.

```python
dataset.head()
```

# To remove null values

```python
dataset.dropna(inplace = True)
```

# Task4: In which month do people book Uber rides less Frequently?

```python
dataset.head()
```

# we will add one more column MONTH by extracting from START_DATE

```python
dataset['MONTH'] = pd.DatetimeIndex(dataset['START_DATE']).month

month_label  = {1.0: 'Jan', 2.0: 'Feb', 3.0: 'Mar', 4.0: 'April',
               5.0: 'May', 6.0: 'June', 7.0: 'JUly', 8.0 : 'Aug',
               9.0: 'Sep', 10.0: 'oct', 11.0: 'Nov', 12.0:'Dec'}  # Month ko string name assign karta hai

dataset['MONTH']  = dataset.MONTH.map(month_label) # Number month ko string names se replace karta h
mon = dataset.MONTH.value_counts(sort = False)  # har month k count calculate karta hai
```

```python
dataset.head()
```

```python
df = pd.DataFrame({
    'MONTH': mon.values, # har month ka total count
    'VALUE COUNT': dataset.groupby('MONTH', sort =False)['MILES'].max() # har month ke max MILES ka value.
})

p = sns.lineplot(data = df) # LIne plot banata hai
p.set(xlabel = 'MONTH',ylabel = 'VALUE COUNT') # Axis labels set krta hai

plt.show()
```

# CONCLUSION - Jan,Feb,Nov & Dec months people book Uber rides less Frequently.

# Task5: On which days of the week do people book uber rides most?

# we will add one more column DAY by extracting from START_DATE

```python
dataset['DAY'] = dataset.START_DATE.dt.weekday

day_label = {
       0:'Mon', 1: 'Tues', 2: 'Wed', 3: 'Thru', 4: 'Fri',
       5: 'Sat', 6: 'Sat'}
dataset['DAY'] = dataset['DAY'].map(day_label)
```

```python
dataset.head()
```

```python
day_label = dataset.DAY.value_counts()

sns.barplot(x = day_label.index,y = day_label)
plt.xlabel('DAY')
plt.ylabel('COUNT')
plt.show()
```

# CONCLUSION -  FRIDAY of the week do people book uber rides most.
